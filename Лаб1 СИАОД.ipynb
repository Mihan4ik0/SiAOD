```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.

Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def selection_sort(arr): 
    for i in range(len(arr)): 
        min_idx = i 
        for j in range(i+1, len(arr)): 
            if arr[min_idx] > arr[j]: 
                min_idx = j 
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
for row in range(len(Matrix)):
     selection_sort(Matrix[row])
sort_arr = selection_sort(Matrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
# Сортировка вставкой.
Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def insertion_sort(arr): 
    for i in range(1, len(arr)): 
        key = arr[i] 
        j = i-1
        while j >=0 and key < arr[j] : 
                arr[j+1] = arr[j] 
                j -= 1
        arr[j+1] = key 
    return arr
for row in range(len(Matrix)):
     insertion_sort(Matrix[row]) 
sort_arr = insertion_sort(Matrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
# Сортировка обменом. || Сортировка пузырьком.
Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def bubble_sort(arr): 
    n = len(arr) 
    for i in range(n): 
        for j in range(0, n-i-1): 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
    return arr
for row in range(len(Matrix)):
     bubble_sort(Matrix[row]) 
sort_arr = bubble_sort(Matrix) 
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
# Сортировка Шелла.
Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def shell_sort(arr): 
    n = len(arr) 
    gap = n//2
    while gap > 0: 
        for i in range(gap,n): 
            temp = arr[i] 
            j = i 
            while  j >= gap and arr[j-gap] >temp: 
                arr[j] = arr[j-gap] 
                j -= gap 
            arr[j] = temp 
        gap //= 2
    return arr
for row in range(len(Matrix)):
     insertion_sort(Matrix[row]) 
sort_arr = shell_sort(Matrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
# Быстрая сортировка.
Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def quick_sort(mas):
    quick_sort_helper(mas, 0, len(mas) - 1)
    return mas


def quick_sort_helper(mas, low, high):
    if low < high:
        split_point = partition(mas, low, high)
        quick_sort_helper(mas, low, split_point - 1)
        quick_sort_helper(mas, split_point + 1, high)


def partition(mas, low, high):
    pivot_value = mas[low]
    left_mark = low + 1
    right_mark = high
    done = False

    while not done:
        while left_mark <= right_mark and mas[left_mark] <= pivot_value:
            left_mark += 1
        while right_mark >= left_mark and mas[right_mark] >= pivot_value:
            right_mark -= 1

        if right_mark < left_mark:
            done = True
        else:
            mas[left_mark], mas[right_mark] = mas[right_mark], mas[left_mark]

    mas[low], mas[right_mark] = mas[right_mark], mas[low]
    return right_mark
for row in range(len(Matrix)):
     quick_sort(Matrix[row])
sort_arr = quick_sort(Matrix)   
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
# Турнирная сортировка.
Matrix = [ [ random.randint(user_min_limit, user_max_limit) for j in range(user_n)] for i in range(user_m) ]
for i in range(user_m):
  print(Matrix[i])
start_time = time.time()
def tournament_sort(arr): 
    if len(arr) > 1: 
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:] 
        tournament_sort(left) 
        tournament_sort(right) 
        i = 0
        j = 0
        k = 0
        while i < len(left) and j < len(right): 
            if left[i] < right[j]: 
                arr[k]=left[i] 
                i+=1
            else: 
                arr[k]=right[j] 
                j+=1
            k+=1
        while i < len(left): 
            arr[k]=left[i] 
            i+=1
            k+=1
        while j < len(right): 
            arr[k]=right[j] 
            j+=1
            k+=1
    return arr
for row in range(len(Matrix)):
     tournament_sort(Matrix[row]) 
sort_arr = tournament_sort(Matrix)   
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for i in range(user_m):
  print(sort_arr[i])
```

```python
def std_sort(arr):
    arr.sort()
```

```python
import time
import random
import matplotlib.pyplot as plt
N = [100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000]
T_selection = []
T_insertion = []
T_bubble = []
T_shell = []
T_quick = []
T_tournament = []
T_standard = []
for n in N:
    arr = [random.randint(0, n) for _ in range(n)]
    # Selection Sort
    start = time.time()
    selection_sort(arr)
    end = time.time()
    T_selection.append(end-start)
    # Insertion Sort
    start = time.time()
    insertion_sort(arr)
    end = time.time()
    T_insertion.append(end-start)
    # Bubble Sort
    start = time.time()
    bubble_sort(arr)
    end = time.time()
    T_bubble.append(end-start)
    # Shell Sort
    start = time.time()
    shell_sort(arr)
    end = time.time()
    T_shell.append(end-start)
    # Quick Sort
    start = time.time()
    quick_sort(arr)
    end = time.time()
    T_quick.append(end-start)
    # Tournament Sort
    start = time.time()
    tournament_sort(arr)
    end = time.time()
    T_tournament.append(end-start)
    # Standard Sort
    start = time.time()
    std_sort(arr)
    end = time.time()
    T_standard.append(end-start)
# Plotting
plt.plot(N, T_selection, label="Selection Sort")
plt.plot(N, T_insertion, label="Insertion Sort")
plt.plot(N, T_bubble, label="Bubble Sort")
plt.plot(N, T_shell, label="Shell Sort")
plt.plot(N, T_quick, label="Quick Sort")
plt.plot(N, T_tournament, label="Tournament Sort")
plt.plot(N, T_standard, label="Standard Sort")
plt.xlabel('Size of matrix')
plt.ylabel('Time (in sec)')
plt.title('Sorting Algorithm Comparison')
plt.legend()
plt.show()
```
