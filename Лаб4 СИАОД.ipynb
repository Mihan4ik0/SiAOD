```python
# Бинарный поиск
def binary_search(lst, x):
    # Устанавливаем начальные значения для нижней и верхней границ
    low = 0
    high = len(lst) - 1
    # Пока нижняя граница не сравняется или пересечется с верхней границей
    while low <= high:
        # Находим середину списка
        mid = (low + high) // 2
        # Если элемент найден в середине списка, возвращаем его индекс
        if lst[mid] == x:
            return mid
        # Если элемент меньше значения середины, изменяем верхнюю границу
        elif lst[mid] < x:
            low = mid + 1
        # Если элемент больше значения середины, изменяем нижнюю границу
        else:
            high = mid - 1

    # Если элемент не найден в списке, возвращаем None
    return None
```

```python
# Бинарное дерево

class Node:
    def __init__(self, val=None, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class BinaryTree:
    def __init__(self, lst):
        # Инициализация бинарного дерева из отсортированного списка
        self.root = self.buildtree(lst, 0, len(lst)-1)

    def buildtree(self, lst, start, end):
        # Рекурсивное построение дерева
        if start > end:
            # Базовый случай: отрезок пуст, возвращаем None
            return None
        mid = (start + end) // 2
        root = Node(val=mid)
        root.left = self.buildtree(lst, start, mid-1)
        root.right = self.buildtree(lst, mid+1, end)
        return root

    def search(self, x):
        # Поиск узла с заданным значением
        node = self.root
        while node is not None:
            if node.val == x:
                return node
            elif node.val < x:
                node = node.right
            else:
                node = node.left
        return None
```

```python
# Поиск фибоначчи
def fibonacci_search(lst, x):
    # Инициализация переменных Фибоначчи
    fib2 = 0
    fib1 = 1
    fib = fib1 + fib2
    
    # Нахождение ближайшего числа Фибоначчи к длине списка
    while fib < len(lst):
        fib2 = fib1
        fib1 = fib
        fib = fib1 + fib2
        
    offset = -1
    
    # Поиск значения x в списке
    while fib > 1:
        i = min(offset+fib2, len(lst)-1)
        if lst[i] < x:
            fib, fib1, fib2 = fib1, fib2, fib1-fib2
            offset = i
        elif lst[i] > x:
            fib, fib1, fib2 = fib2, fib1-fib2, fib2-fib1
        else:
            return i
            
    # Проверка последнего элемента и возврат результата
    if fib1 and lst[offset+1] == x:
        return offset+1
    return None
```

```python
# Интерполяционный
def interpolation_search(lst, x):
    # Определяем начальные значения для поиска
    low = 0
    high = len(lst) - 1
    # Ищем элемент, пока он не будет найден или пока список не будет полностью просмотрен
    while low <= high and lst[low] <= x <= lst[high]:
        # Определяем индекс среднего элемента для текущего диапазона поиска
        mid = low + int(((x - lst[low]) * (high - low)) / (lst[high] - lst[low]))
        
        # Если элемент найден, возвращаем его индекс
        if lst[mid] == x:
            return mid
        # Если искомый элемент находится в правой половине, продолжаем поиск в ней
        elif lst[mid] < x:
            low = mid + 1
        # Если искомый элемент находится в левой половине, продолжаем поиск в ней
        else:
            high = mid - 1

    # Если элемент не был найден, возвращаем None
    return None
```

```python
# Простое рехэширование
class LinearProbingHashTable:
    def __init__(self, size):
        self.size = size # размер хэш-таблицы
        self.keys = [None] * size # массив ключей (изначально заполняем None)
        self.values = [None] * size # массив значений (изначально заполняем None)

    def _hash(self, key): # хэш-функция
        return key % self.size

    def insert(self, key, value): # метод для добавления ключа и значения
        i = self._hash(key)

        # пока следующая ячейка не пуста и ключ не совпадает, перемещаемся вправо
        while self.keys[i] is not None and self.keys[i] != key:
            i = (i + 1) % self.size

        # добавляем пару ключ-значение в найденную ячейку
        self.keys[i] = key
        self.values[i] = value

    def search(self, key): # метод для поиска значения по ключу
        i = self._hash(key)

        # пока ячейка не пуста
        while self.keys[i] is not None:
            if self.keys[i] == key: # если ключ совпал
                return self.values[i] # возвращаем значение
            i = (i + 1) % self.size # перемещаемся вправо

        return None # если ключ не найден, возвращаем None

    def delete(self, key): # метод для удаления элемента по ключу
        i = self._hash(key)

        while self.keys[i] is not None: # пока ячейка не пуста
            if self.keys[i] == key: # если ключ совпал
                self.keys[i] = None # удаляем ключ
                self.values[i] = None # удаляем значение

                j = (i + 1) % self.size # начинаем поиск с следующей ячейки
                while self.keys[j] is not None: # пока не дойдем до пустой ячейки
                    k = self._hash(self.keys[j]) # вычисляем хэш для следующего ключа
                    if k != j: # если новый индекс не совпадает с текущим
                        # перемещаем пару ключ-значение на правильную позицию
                        self.keys[j], self.values[j] = self.keys[k], self.values[k]
                        self.keys[k], self.values[k] = None, None
                    j = (j + 1) % self.size # перемещаемся вправо

                return # элемент удален

            i = (i + 1) % self.size # перемещаемся вправо
```

```python
# Рехэширование с помощью псевдослучайных чисел
class RandomProbingHashTable:
    def __init__(self, size):
        # Инициализация размера хеш-таблицы и списков ключей и значений.
        self.size = size
        self.keys = [None] * size
        self.values = [None] * size
        
    # Вспомогательная функция, которая возвращает индекс для ключа в зависимости от количества попыток вставки.
    def _hash_function(self, key, attempt):
        return (hash(key) + attempt * random.randint(1, 1001)) % self.size
        
    # Метод вставки ключа и значения в хеш-таблицу.
    def put(self, key, value):
        for i in range(self.size):
            # Получаем индекс для текущей попытки вставки.
            index = self._hash_function(key, i)
            # Если ячейка с данным индексом пуста, вставляем ключ и значение.
            if self.keys[index] == None:
                self.keys[index] = key
                self.values[index] = value
                return 
            # Если ячейка содержит ключ, равный добавляемому, обновляем значение.
            elif self.keys[index] == key:
                self.values[index] = value
                return 
        # Если вставка не удалась после всех попыток, вызываем исключение.
        raise Exception("Hash table is full")
        
    # Метод поиска значения по ключу в хеш-таблице.
    def get(self, key):
        for i in range(self.size):
            # Получаем индекс для текущей попытки поиска.
            index = self._hash_function(key, i)
            # Если ячейка с данным индексом пуста, значит, ключа нет в хеш-таблице.
            if self.keys[index] == None:
                return None
            # Если ячейка содержит ключ, равный искомому, возвращаем соответствующее значение.
            elif self.keys[index] == key:
                return self.values[index]
        # Если ключ не был найден после всех попыток, возвращаем None.
        return None
```

```python
class ChainedHashTable:
    def __init__(self, size):
        self.size = size
        # Создание пустой таблицы фиксированного размера
        self.table = [[] for _ in range(size)]

    def _hash(self, key):
        # Возвращает индекс в таблице для указанного ключа
        return key % self.size

    def put(self, key, value):
        # Добавление значения в таблицу по указанному ключу
        hash_value = self._hash(key)
        # Поиск элемента в таблице с ключом key
        for i, (k, v) in enumerate(self.table[hash_value]):
            # Если элемент с ключом key уже существует, обновляем его значение
            if k == key:
                self.table[hash_value][i] = (key, value)
                return
        # Если элемента с ключом key нет, добавляем его в таблицу
        self.table[hash_value].append((key, value))

    def get(self, key):
        # Получение значения из таблицы по указанному ключу
        hash_value = self._hash(key)
        # Поиск элемента в таблице с ключом key
        for k, v in self.table[hash_value]:
            # Если элемент с ключом key найден, возвращаем его значение
            if k == key:
                return v
        # Если элемента с ключом key нет, возвращаем None
        return None
```

```python
x = random.randint(0, 100000)
lst = sorted(random.sample(range(100000), 100000))
print(f"Search for {x} in list of size {len(lst)}")
print("Binary search: ", timeit.timeit(lambda: binary_search(lst, x), number=100)  * 1000, 'ms')
print("Binary tree search: ", timeit.timeit(lambda: BinaryTree(lst).search(x), number=100) * 1000, 'ms')
print("Fibonacci search: ", timeit.timeit(lambda: fibonacci_search(lst, x), number=100) * 1000, 'ms')
print("Interpolation search: ", timeit.timeit(lambda: interpolation_search(lst, x), number=100) * 1000, 'ms')
print("Simple rehashing hash table search: ", timeit.timeit(lambda: LinearProbingHashTable(len(lst)).search(x), number=100) * 1000, 'ms')
print("Random rehashing hash table search: ", timeit.timeit(lambda: RandomProbingHashTable(len(lst)).get(x), number=100) * 1000, 'ms')
print("Chained hash table search: ", timeit.timeit(lambda: ChainedHashTable(len(lst)).get(x), number=100) * 1000, 'ms')
print("Python built-in search: ", timeit.timeit(lambda: lst.index(x), number=100) * 1000, 'ms')
```

# Задание №3

Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
def print_board(board):
    n = len(board)
    print("  " + " ".join([chr(ord('a') + i) for i in range(n)]))
    for i in range(n):
        print(str(n - i) + " " + " ".join(board[i]))
    print()


# Определим функцию для проверки, можно ли разместить ферзя в данной позиции на доске.
def is_safe(board, row, col, n):
    # Проверяем, есть ли ферзь в той же строке.
    for i in range(col):
        if board[row][i] == "Q":
            return False

    # Проверяем, есть ли ферзь в верхней левой диагонали.
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == "Q":
            return False

    # Проверяем, есть ли ферзь в нижней левой диагонали.
    for i, j in zip(range(row, n, 1), range(col, -1, -1)):
        if board[i][j] == "Q":
            return False

    # Если все проверки пройдены, то размещение ферзя в данной позиции безопасно.
    return True

# Определим функцию для решения задачи n-ферзей с помощью метода возврата.
def solve_n_queens(board, col, n):
    # Базовый случай: все столбцы заполнены ферзями.
    if col >= n:
        return True

    # Попробуем разместить ферзя в каждой строке текущего столбца, начиная с верхней.
    for i in range(n):
        if is_safe(board, i, col, n):
            board[i][col] = "Q"

            # Рекурсивно решаем подзадачу для следующего столбца.
            if solve_n_queens(board, col + 1, n):
                return True

            # Возвращаемся к предыдущему шагу и удаляем ферзя из текущей позиции.
            board[i][col] = "."

    # Если решение не найдено, возвращаем False.
    return False

# Определим главную функцию.
def main():
    n = 8
    board = [["." for _ in range(n)] for _ in range(n)]

    # Решаем задачу n-ферзей и выводим решение, если оно существует.
    if solve_n_queens(board, 0, n):
        print_board(board)
    else:
        print("Решения не существует.")

# Вызываем главную функцию, если скрипт запущен напрямую.
if __name__ == "__main__":
    main()
```
